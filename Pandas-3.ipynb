# Merge() Function

```python
df1 = pd.DataFrame({'Roll no' : [1,2,3,4,5],'Physics':[34,67,34,89,12]})
df1
```

```python
df2 = pd.DataFrame({'Roll no' : [1,2,3,4,5],'Chemistry':[11,12,13,14,15]})
df2
```

```python
pd.merge(df1,df2,on = 'Roll No')
```

```python
pd.merge(df2,df1)
```

```python
pd.merge(df1,df2)
```

```python
df3 = pd.DataFrame({'Roll no' : [1,2,3,6,7],'Physics':[34,67,34,89,12]})
df3
df4 = pd.DataFrame({'Roll no' : [1,2,3,8,9],'Chemistry':[34,67,34,89,12]})
df4
```

```python
pd.merge(df3,df4)
```

```python
pd.merge(df3,df4,how = 'left')
```

```python
pd.merge(df3,df4,how = "right")
```

```python
pd.merge(df3,df4,how = 'outer')
```

# Common Aggregation Functions:
# 'mean': Computes the mean (average) of numeric values.
# 'sum': Sums up numeric values.
# 'min': Finds the minimum value.
# 'max': Finds the maximum value.
# 'count': Counts the number of occurrences (non-null values).
# 'median': Computes the median of numeric values.
# 'var': Computes the variance.
# 'first': Returns the first value in each group.
# 'prod': Returns the product of the numeric values.
# 'mad': Mean absolute deviation.
# 'nunique': Counts the number of unique values.
